```python
def type_mapper(string: str) -> str:
    string = string.lower()

    if "int" in string or "bool" in string:
        result = "bigint"

    elif "date" in string:
        result = "date"

    elif "float" in string:
        result = "decimal(32,8)"

    else:
        result = "varchar(32)"

    return result.upper()
```

```python
def generate_create_query(df: pd.DataFrame, schema: str, table_name: str) -> str:

    df_meta = pd.DataFrame(df.dtypes).reset_index().astype(str)
    df_meta.columns = ["columnname", "python_dtype"]

    df_meta.loc[:, "teradata_dtype"] = df_meta.python_dtype.str.lower().map(
        lambda x: type_mapper(x)
    )
    print(df_meta)

    primary_index = ", ".join(
        [x for x in (df_meta.columnname) if "id" in x or "class" in x]
    )
    print(f"\nPrimary Index: {primary_index}")

    query = (
        f"create set table {schema}.{table_name} , no fallback ("
        + ", ".join(list(df_meta.columnname + " as " + df_meta.teradata_dtype))
        + f") primary index({primary_index});"
    )
    return query
```

```python
test1_df = pd.DataFrame({"value": np.random.logistic(size=5_000)})

test1_df["group"] = (
    pd.Series((["a", "b", "c", "d", "e"] * 1_000))
    .sample(frac=1)
    .reset_index()
    .loc[:, 0]
)
```

```python
test2_df = pd.DataFrame({"value": np.random.logistic(size=5_000)})

test2_df["group"] = (
    pd.Series((["a", "b", "c", "d", "e"] * 1_000))
    .sample(frac=1)
    .reset_index()
    .loc[:, 0]
)
```

```python
test2_df
```

```python
test1_df
```

```python
_groupby_ = test1_df.groupby(["group"])
_temp1_df_ = pd.DataFrame(
    {
        "mean": _groupby_.value.mean(),
        "median": _groupby_.value.median(),
        "count": _groupby_.value.count(),
    }
).reset_index()

_groupby_ = test2_df.groupby(["group"])
_temp2_df_ = pd.DataFrame(
    {
        "mean": _groupby_.value.mean(),
        "median": _groupby_.value.median(),
        "count": _groupby_.value.count(),
    }
).reset_index()
```

```python
_temp1_df_
```

```python
from scipy.stats import wilcoxon, mannwhitneyu
```

```python
pvalue_df = pd.DataFrame()

for x in _temp1_df_.group.drop_duplicates():
    _series_1_ = test1_df.loc[test1_df.group == x, "value"]
    _series_2_ = test2_df.loc[test2_df.group == x, "value"]

    res = wilcoxon(x=_series_1_, y=_series_2_)
    res_mann = mannwhitneyu(x=_series_1_, y=_series_2_)

    sub_df = pd.DataFrame(
        {
            "group": [x],
            "p_value_wilcoxon": [res.pvalue],
            "p_value_mann": [res_mann.pvalue],
        }
    )

    pvalue_df = pd.concat([pvalue_df, sub_df])

pvalue_df = pvalue_df.reset_index().drop(columns=["index"])
```

```python
result_df = _temp1_df_
for _df_ in [_temp2_df_, pvalue_df]:
    result_df = pd.merge(
        left=result_df, right=_df_, left_on="group", right_on="group", how="outer"
    )

result_df.sort_values("p_value_mann", ascending=True)
```

```python
r = np.random.RandomState(seed=5)
ints = r.random_integers(1, 10, size=(3, 2))
a = ["A", "B", "C"]
b = ["D", "A", "E"]
df = pd.DataFrame(ints, columns=["weight", "cost"])
df[0] = a
df["b"] = b
df
```

```python
G = nx.from_pandas_edgelist(df=df, source=0, target="b")
```

```python
for edge in G.edges():
    print(G.nodes[edge[0]])
```

```python
edge_x = []
edge_y = []
for edge in G.edges():
    x0, y0 = G.nodes[edge[0]]["pos"]
    x1, y1 = G.nodes[edge[1]]["pos"]
    edge_x.append(x0)
    edge_x.append(x1)
    edge_x.append(None)
    edge_y.append(y0)
    edge_y.append(y1)
    edge_y.append(None)
```

```python
edge_trace = go.Scatter(
    x=edge_x,
    y=edge_y,
    line=dict(width=1.0, color="#888"),
    hoverinfo="none",
    mode="lines",
)
```

```python
node_x = []
node_y = []
for node in G.nodes():
    x, y = G.nodes[node]["pos"]
    node_x.append(x)
    node_y.append(y)

node_trace = go.Scatter(
    x=node_x,
    y=node_y,
    mode="markers",
    hoverinfo="text",
    # text=[x[0] for x in G.adjacency()],
    marker=dict(
        showscale=True,
        # colorscale options
        #'Greys' | 'YlGnBu' | 'Greens' | 'YlOrRd' | 'Bluered' | 'RdBu' |
        #'Reds' | 'Blues' | 'Picnic' | 'Rainbow' | 'Portland' | 'Jet' |
        #'Hot' | 'Blackbody' | 'Earth' | 'Electric' | 'Viridis' |
        colorscale="YlGnBu",
        reversescale=True,
        color=[],
        size=50,
        colorbar=dict(
            thickness=15, title="Node Connections", xanchor="left", titleside="right"
        ),
        line_width=2,
    ),
)
```

```python
node_adjacencies = []
node_text = []
for node, adjacencies in enumerate(G.adjacency()):
    node_adjacencies.append(len(adjacencies[1]))
    node_text.append(f"{adjacencies[0]} # of connections: " + str(len(adjacencies[1])))

node_trace.marker.color = node_adjacencies
node_trace.text = node_text
```

```python
fig = go.Figure(
    data=[edge_trace, node_trace],
    layout=go.Layout(
        title="<br>Network graph made with Python",
        titlefont_size=16,
        showlegend=False,
        hovermode="closest",
        margin=dict(b=20, l=5, r=5, t=40),
        annotations=[
            dict(
                text="Python code: <a href='https://plotly.com/ipython-notebooks/network-graphs/'> https://plotly.com/ipython-notebooks/network-graphs/</a>",
                showarrow=False,
                xref="paper",
                yref="paper",
                x=0.005,
                y=-0.002,
            )
        ],
        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    ),
)
fig.show(renderer="browser")
```
